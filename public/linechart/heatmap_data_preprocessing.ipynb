```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
def anamoly_process(file):

    main_data=pd.read_csv("/Users/raghav/Desktop/VAST-2017-MC2-Venkatramanan/MC2Data/"+ file +"_new.csv")
    data_new=pd.read_csv("/Users/raghav/Desktop/VAST-2017-MC2-Venkatramanan/linechart/data/"+ file +"_new.csv")

    for i, j in data_new.iterrows(): 
        data_new["date"][i] = datetime.strptime(j.date, "%m/%d/%Y %H:%M")


    m = []
    c = []
    a = []
    ap = []

    for i, j in main_data.iterrows(): 
        if j.anomaly_count == 'a' and j.chemical == "Methylosmolene":
            m.append(j.date)
        elif j.anomaly_count == 'a' and j.chemical == "AGOC-3A":
            a.append(j.date)
        elif j.anomaly_count == 'a' and j.chemical == "Chlorodinine":
            c.append(j.date)
        elif j.anomaly_count == 'a' and j.chemical == "Appluimonia":
            ap.append(j.date)


    data_new["Methylosmolene_anamoly"] = np.nan
    data_new["Chlorodinine_anamoly"] = np.nan
    data_new["AGOC_anamoly"] = np.nan
    data_new["Appluimonia_anamoly"] = np.nan

    for i, j in data_new.iterrows(): 
        if str(j.date) in m:
            data_new["Methylosmolene_anamoly"][i] = 1
        elif str(j.date) in a:
            data_new["AGOC_anamoly"][i] = 1
        elif str(j.date) in c:
            data_new["Chlorodinine_anamoly"][i] = 1
        elif str(j.date) in ap:
            data_new["Appluimonia_anamoly"][i] = 1
    
    data_new["Methylosmolene_anamoly"] = data_new["Methylosmolene_anamoly"].fillna(0)
    data_new["AGOC_anamoly"] = data_new["AGOC_anamoly"].fillna(0)
    data_new["Chlorodinine_anamoly"] = data_new["Chlorodinine_anamoly"].fillna(0)
    data_new["Appluimonia_anamoly"] = data_new["Appluimonia_anamoly"].fillna(0)
    
    data_new.to_csv("/Users/raghav/Desktop/VAST-2017-MC2-Venkatramanan/MC2Data/"+ file +"_new_anamoly.csv")
```

```python
anamoly_process("sensor_1")
anamoly_process("sensor_2")
anamoly_process("sensor_3")
anamoly_process("sensor_4")
anamoly_process("sensor_5")
anamoly_process("sensor_6")
anamoly_process("sensor_7")
anamoly_process("sensor_8")
anamoly_process("sensor_9")
```
